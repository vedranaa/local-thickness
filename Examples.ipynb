# Local thickness in 2D and 3D

**Minimal working example for 2D and 3D local thickness using `localthickness` module.** 

Authors: abda@dtu.dk, vand@dtu.dk

For an object in 2D or 3D, the local thickness in any point of the object is defined as the radius of the larges sphere which fits inside the object and contains the point. 

Function for the computation of local thickness takes a binary 2D or 3D image as input, where ones represent the object to be measured. The function returns an array of the same size as the input which in every pixel (voxel) carries information about the local thickness of the object.

Function for fast computation of local thickness utilizes two approximations for a more effecient computation. 
* First, the dilation with the sphere of a large radius is approximated with the consecutive dilations with the sphere of radius one. The results of such a dilation is not perfectly spherical (it is polygonal/polyhedral), but it provides a good approximation. 
* Second, the function may use a scaled approach, with local thickness computed for a downscaled image image first, and then upscaled.

Function call is
```
T = local_thickness(B, mask=None, scale=1)
```
Here, `B` is a binary image, `mask` is a binary mask of the same size as `B` and `scale` is downscaling factor (smaller than 1).

## Local thickness in 2D
The 2D example loads a synthetic test image and computes local thickness.

````python
import skimage.io
import matplotlib.pyplot as plt
import localthickness as lt

# Load image with test objects to be analysed and make it binary.
B = skimage.io.imread('data/test_2D_A.png', as_gray = True) > 0.5 

# Compute local thickness using scaled approach.
T = lt.local_thickness(B, scale=0.25)

# Visualize.
fig, ax = plt.subplots(1, 3, figsize=(10,5), gridspec_kw={"width_ratios":[1, 1, 0.05]})
ax[0].imshow(B, cmap=plt.cm.gray)
ax[0].set_title('Test objects')
h = ax[1].imshow(T, cmap=lt.black_plasma())
ax[1].set_title('Local thickness')
fig.colorbar(h, cax=ax[2])
plt.show()
````

## Fast local thickness in 3D
We compute local thickness and show it on a few slices from the volume. 

````python
import matplotlib.pyplot as plt
import localthickness as lt
import tifffile

# Load volume with test objects to be analysed and make it binary.
B = tifffile.imread('data/cotton_test_volume.tif') > 0

# Compute local thickness using scaled approach.
T = lt.local_thickness(B, scale=0.25)

# Visualize slices of the result.
fig, ax = plt.subplots(2, 6, figsize=(10,4), sharex='col', sharey='col', 
                       gridspec_kw={"width_ratios":[1, 1, 1, 1, 1, 0.05]})
vmax = T.max()
for i in range(5):
    z = ((i+2)*(B.shape[0]))//8
    h0 = ax[0][i].imshow(B[z], cmap=plt.cm.gray)
    ax[0][i].set_title(f'Slice {z}')
    h1 = ax[1][i].imshow(T[z], vmax=vmax, cmap=lt.black_plasma())
    ax[1][i].set_title(f'Slice {z}')
fig.colorbar(h1, cax=ax[1][5]) 
ax[0][5].remove()
plt.show()
````

In the visualization below you can use keybord keys (`left`/`a`, `right`/`d`, `up`/`w`, `down`/`z`, `pageup`/`e`, and `pagedown`/`x`) to inspect the slices of the volume.
````python
# visulizing slices of the volume
%matplotlib notebook
import localthickness as lt
lt.show_vol(T, cmap=lt.black_plasma())
````

With [plotly](https://plot.ly/python/) we can see the data and the thickness as an interactive 3D visualization. __Be patient__, tt may take up to a minute for plotly to produce a figure. 

````python
import plotly.graph_objects as go
import numpy as np

X, Y, Z = np.mgrid[range(B.shape[0]), range(B.shape[1]), range(B.shape[2])]

fig = go.Figure(data = go.Volume(x=X.flatten(), y=Y.flatten(), z=Z.flatten(), value=T.flatten(),
    isomin=0, isomax=T.max(), opacity=0.1, surface_count=15, colorscale = lt.pl_black_plasma()))
fig.show()
````
